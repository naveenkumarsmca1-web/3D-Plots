```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
total_points=1000
```

```python
x=np.random.uniform(-400,400,total_points)
y=np.random.uniform(-400,400,total_points)
z=np.random.uniform(-400,400,total_points)
print(x[:3])
```

```python
points=[]
for num in range(total_points):
    points.append({"x":float(x[num]),
                   "y":float(y[num]),
                   "z":float(z[num])})
```

```python

print(points[:4])
```

```python
with open("data.json","w")as file:
    data=json.dump(points,file,indent=4)
print(f"{total_points}data generated successfully")
```
